```python
close = pd.DataFrame(data=np.random.random_sample(100), columns=["price"])
sma5 = close.rolling(window=5).mean().fillna(method="bfill")
sma5.columns = ["sma5"]
sma20 = close.rolling(window=20).mean().fillna(method="bfill")
sma20.columns = ["sma20"]
label = pd.DataFrame(data=np.zeros(shape=close.shape[0]), index=close.index, columns=["direction"])
```

```python
for j, val in label.direction.iteritems():
    if sma5.iloc[:,0].loc[j] >= sma20.iloc[:,0].loc[j]:
        label.direction.loc[j] = 1
    if sma5.iloc[:,0].loc[j] < sma20.iloc[:,0].loc[j]:
        label.direction.loc[j] = -1
plt.plot(sma5)
plt.plot(sma20)
plt.plot(label)
```
